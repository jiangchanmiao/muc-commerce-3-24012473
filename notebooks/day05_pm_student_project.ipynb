## 任务3：单维专题分析（必做）

根据所选专题确定一个主分组字段，并使用`groupby + agg`完成命名聚合。

最低要求：

- 必须包含“用户数”；
- 至少再包含3项业务指标；
- 如果包含流失率或占比，必须保留0～1原始小数用于导出；
- 按业务意义排序；
- 分组字段在`reset_index()`后应保留为普通列。

```python
topic_fields = {
    "A": {"TenureGroup"},
    "B": {"Complain", "SatisfactionScore"},
    "C": {"PreferedOrderCat"},
    "D": {"PreferredPaymentMode"},
    "E": {"CityTier", "PreferredLoginDevice"},
}

print("当前专题：", TOPIC)
print("可选主分组字段：", topic_fields[TOPIC])


# 选择主分组字段
segment_field = "PreferredPaymentMode"


# 使用groupby + agg完成命名聚合
segment_analysis = (
    df.groupby(segment_field, observed=True)
    .agg(
        用户数=("CustomerID", "count"),
        流失率=("Churn", "mean"),
        平均订单数=("OrderCount", "mean"),
        平均优惠券数=("CouponUsed", "mean"),
        平均返现=("CashbackAmount", "mean"),
    )
    .reset_index()
)


# 重置索引、按业务意义排序并展示
segment_analysis = segment_analysis.sort_values("用户数", ascending=False).reset_index(drop=True)
display(segment_analysis)
```

```python
# 检查点3：单维专题分析

assert segment_field in df.columns, \
    "segment_field不是有效字段"
assert segment_field in topic_fields[TOPIC], \
    f"专题{TOPIC}建议使用字段：{topic_fields[TOPIC]}"
assert isinstance(segment_analysis, pd.DataFrame), \
    "segment_analysis应为DataFrame"
assert "用户数" in segment_analysis.columns, \
    "专题分析表必须包含用户数"
assert len(segment_analysis) >= 2, \
    "专题分析至少应包含两个分组"
assert segment_analysis["用户数"].sum() == len(df), \
    "各分组用户数之和应等于总用户数"

print("检查点3通过")
```

### 单维专题分析记录

**本专题要回答的业务问题：**

> 不同支付偏好用户的规模和流失率、优惠行为是否存在分组差异？

**数据现象：**

> 在 5 种支付偏好中，DebitCard用户数最多（2,314人，占41.1%），流失率为15.38%，平均订单数为2.94，平均优惠券使用次数为1.72，平均返现金额为177.06元；CreditCard用户为1,774人，流失率最低（14.21%），平均订单数为3.05；E-wallet用户为614人，流失率为22.80%，显著高于总体流失率16.84%；COD用户为514人，流失率最高达24.90%，但平均优惠券使用次数为1.82，反而略高于其他支付方式；UPI用户为414人，流失率为17.39%，平均订单数最低（2.57）。COD和E-wallet用户的流失率均明显高于总体水平。

**可能解释：**

> 支付偏好与流失率之间可能存在关联，COD 用户流失率较高值得关注，这可能与用户支付习惯、优惠行为等相关。但支付偏好并非唯一影响因素，需结合城市层级、满意度等变量进一步验证，不能据此判定支付方式直接导致流失。

## 任务4：双维度交叉分析（必做）

从以下维度中选择两个不同字段：

- `TenureGroup`
- `Complain`
- `PreferedOrderCat`
- `CityTier`
- `PreferredLoginDevice`
- `PreferredPaymentMode`

最低要求：

- 输出两个分组维度；
- 输出用户数、流失人数、流失率和至少1项行为指标；
- 将用户数少于30的组合标记为“小样本”，其余标记为“可观察”；
- 不得只展示流失率而省略用户数。

```python
allowed_cross_fields = {
    "TenureGroup",
    "Complain",
    "PreferedOrderCat",
    "CityTier",
    "PreferredLoginDevice",
    "PreferredPaymentMode",
}


# 选择两个不同维度
dim_1 = "PreferredPaymentMode"
dim_2 = "CityTier"


# 使用groupby + agg完成双维分析
cross_analysis = (
    df.groupby([dim_1, dim_2], observed=True)
    .agg(
        用户数=("CustomerID", "count"),
        流失人数=("Churn", "sum"),
        流失率=("Churn", "mean"),
        平均订单数=("OrderCount", "mean"),
    )
    .reset_index()
)


# 新增"样本提示"列
cross_analysis["样本提示"] = np.where(
    cross_analysis["用户数"] < 30, "小样本", "可观察"
)


# 按流失率排序并展示
cross_analysis = cross_analysis.sort_values("流失率", ascending=False).reset_index(drop=True)
display(cross_analysis)
```

```python
# 检查点4：双维度交叉分析

assert dim_1 in allowed_cross_fields and dim_2 in allowed_cross_fields, \
    "两个分析维度必须来自允许字段"
assert dim_1 != dim_2, "两个分析维度不能相同"
assert isinstance(cross_analysis, pd.DataFrame), \
    "cross_analysis应为DataFrame"

required_cross_cols = {
    dim_1,
    dim_2,
    "用户数",
    "流失率",
    "样本提示",
}

assert required_cross_cols.issubset(cross_analysis.columns), \
    f"双维分析表缺少字段：{required_cross_cols - set(cross_analysis.columns)}"
assert cross_analysis["用户数"].sum() == len(df), \
    "双维组合用户数之和应等于总用户数"
assert set(cross_analysis["样本提示"]).issubset(
    {"小样本", "可观察"}
), "样本提示只能是“小样本”或“可观察”"

expected_sample_hint = np.where(
    cross_analysis["用户数"] < 30,
    "小样本",
    "可观察",
)
assert np.array_equal(
    cross_analysis["样本提示"].to_numpy(),
    expected_sample_hint,
), "样本提示与用户数阈值不一致"

print("检查点4通过")
```

### 双维分析记录

**最值得关注的维度组合：**

> PreferredPaymentMode × CityTier，即"支付偏好 × 城市层级"的组合

**该组合的用户数、流失率和比较对象：**

> COD × CityTier=2 的用户流失率最高，但样本量仅16人，属于小样本。DebitCard × CityTier=1 的用户数最多，流失率接近总体水平。

**是否存在小样本风险：**

> 存在小样本风险。在PreferredPaymentMode与CityTier的14个组合中，仅有一个组合出现小样本：COD × CityTier=2的用户数仅有16人，远低于30人阈值。该组合虽然流失率高达50.00%，但由于样本量过小，属于小样本，可能无法稳定代表总体，解读该组合流失率时需格外谨慎。其余13个组合用户数均≥30，属于可观察范围。

**为什么不能直接写成因果结论：**

> 支付偏好与流失可能只是关联，而非支付方式直接导致流失。存在混杂变量（如满意度、品类偏好等），需要更严格的统计方法才能判断因果关系。
